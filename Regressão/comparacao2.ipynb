```python
import numpy as np
import matplotlib.pyplot as plt

# Carregando resultados
com_hp = np.load('resultados_com_hp_2.npz')
sem_hp = np.load('resultados_sem_hp_2.npz')

# Exemplo para SVR
plt.figure(figsize=(6, 6))
plt.scatter(com_hp['y_test'], com_hp['y_pred_svr'], color='blue', alpha=0.6, label='SVR com hiperparmetros')
plt.scatter(sem_hp['y_test'], sem_hp['y_pred_svr'], color='pink', alpha=0.6, label='SVR sem hiperparmetros')
plt.plot([com_hp['y_test'].min(), com_hp['y_test'].max()], [com_hp['y_test'].min(), com_hp['y_test'].max()], 'k--')
plt.xlabel("Valor Real")
plt.ylabel("Previso")
plt.title("Comparao SVR")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(com_hp['y_test'], com_hp['y_pred_ridge'], color='blue', alpha=0.6, label='ridge com hiperparmetros')
plt.scatter(sem_hp['y_test'], sem_hp['y_pred_ridge'], color='pink', alpha=0.6, label='ridge sem hiperparmetros')
plt.plot([com_hp['y_test'].min(), com_hp['y_test'].max()], [com_hp['y_test'].min(), com_hp['y_test'].max()], 'k--')
plt.xlabel("Valor Real")
plt.ylabel("Previso")
plt.title("Comparao ridge")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
